```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

LOADING DATASET TO PYTHON ENVIRONMENT

```python
df = pd.read_excel("Dataset_w6.xlsx")
df.head()
```

```python
#1. Finding out how many countries are present in the dataset
```

```python
df['country'].count()
```

```python
#2. Examine the data using visualizations
```

```python
#TOP10 COUNTRIES BASED ON POPULATION
```

```python
TOP10 = df.nlargest(n=10, columns=['Population in thousands (2017)'])
TOP10
```

```python
TOP10.plot(x="country", y=["Population in thousands (2017)"], kind="bar",color="RED",width = 0.8, rot=90, fontsize=10,figsize=(7,7))
plt.xlabel('Country')
plt.ylabel('Population in thousands ')
plt.grid()
plt.title('Top 10 Countries Based on Population')
```

```python
#SCATTERPLOT OF REGION VS POPULATION
```

```python
plt.figure(figsize=(20,15))
plt.scatter(df['Region'], df['Population in thousands (2017)'], s=15, color='green')
plt.title('REGION vs POPULATION(in thousands)', fontweight='bold', color='red', fontsize=17)
plt.xlabel('REGION', fontweight='bold')
plt.ylabel('POPULATION (in thousands)', fontweight='bold')
plt.xticks(rotation=90)
plt.show()
```

```python
#HEATMAP SHOWING CORRELATION BETWEEN EACH COLUMNS
```

```python
sns.heatmap(df.corr(),annot=True)
plt.xticks(rotation=45)
```

```python
#BOXPLOT OF GDP PER CAPITA OF REGIONS
```

```python
plt.figure(figsize=(15,20))
sns.boxplot(x='Region', y='GDP per capita (current US$)', data=df)
plt.title('GDP PER CAPITA OF REGIONS', fontweight='bold', fontsize=24, color='VIOLET')
plt.xticks(rotation=75)
plt.show()
```

```python
#Barplot of top 25 countries based on surface area
```

```python
df = pd.read_excel("Dataset_w6.xlsx")
df.head()
plt.figure(figsize=(10,12))
sns.barplot(x='country',y='Surface area (km2)',data=df.sort_values('Surface area (km2)',ascending=False).head(25))
plt.title('Top 25 countries based on Surface area ',fontweight='bold',fontsize=20)
plt.xlabel('Country',fontsize=20)
plt.ylabel('Surface area (km2)',fontsize=15)
plt.xticks(rotation=90)
plt.grid()
plt.show()
```
