```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

data=pd.read_csv('Telco-Customer-Churn.csv')
data

mycolumns= ['Churn','gender']
df=data[mycolumns]
print(df)

sns.countplot(x='gender',hue='Churn',data=df)
plt.title("Gender vs Churn count",fontsize=16)
plt.show()

```

```python
2.H0W MANY FEMALE SENIOR CITIZENS ARE THERE IN THE DATASET
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

data=pd.read_csv('Telco-Customer-Churn.csv')
data

mycolumns= ['gender','SeniorCitizen']
df=data[mycolumns]
print(df)
```

```python
data1=data[data['gender']=='Female']
df=data1[data1['SeniorCitizen']==1]
print("No of female senior citizens  = ",df['SeniorCitizen'].sum())
```

```python
sns.countplot(x=data1['SeniorCitizen'],data=data1)
plt.title('No of Female Senior Citizens')
plt.grid()
plt.show()
```

```python
3.COMPARE TENURE WITH TOTAL CHARGES.
```

```python
plt.figure(figsize=(10,6))
plt.scatter(data['tenure'], data['TotalCharges'],c='red',marker='^', s=8)
plt.title('TENURE VS TOTAL CHARGES', fontsize=15)
plt.xlabel('Tenure')
plt.ylabel('Total Charges')
plt.show()
```

```python
4.FINDOUT WHICH CONTRACT IS PREFERRED BY THE SENIOR CITIZENS.
```

```python
mycolumns= ['SeniorCitizen','Contract']
df=data[mycolumns]
print(df)
```

```python
data_new=df.loc[(data["SeniorCitizen"]==1)]
sns.countplot(x=data_new["Contract"],color='violet')
plt.title("PREFERENCES OF SENIOR CITIZENS")
plt.grid()
plt.show()
```

```python
#plotting preferred payment methods.
plt.figure(figsize=(10,8))
sns.countplot(x=data['PaymentMethod'])
plt.title('PREFERRED PAYMENT METHODS', fontweight='bold', fontsize=18)
plt.xlabel('Payment Method',fontsize='12')
plt.ylabel('Count',fontsize=12)
plt.grid()
plt.show()
```
